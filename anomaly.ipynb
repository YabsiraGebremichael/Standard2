# Data Issues — SBSA DS Case Study 2027.csv

Dataset: **10,000 rows x 36 columns**. All findings below are from a read-only audit of the raw file.

## 1. Duplicates
- **27 duplicated `CustomerID`s** (54 rows involved). Of these:
  - **13 are exact full-row duplicates** (identical data repeated).
  - **14 duplicate IDs have conflicting data** — the same ID appears with different attribute values, so the ID is not a reliable unique key.
- `CustomerID` spans 1–10,000 but only 9,973 are unique.

## 2. Impossible / implausible ages
- **3 negative ages**: -10, -5, -1.
- **107 clients under 18**, including 104 aged 5–17.
- **78 minors (5–17) hold lending/credit products** (Credit_Card, Personal_Loan, Home_Loan and/or Vehicle_Loan) — legally implausible for minors.
- **38 minors are recorded as Married or Divorced**.
- Minors with adult education levels: 16 with Bachelor's, 7 with Postgraduate, 17 with Diploma/Certificate.
- Minors with adult employment: 36 'Full-time employed', 12 'Self-employed', 8 'Retired'.
- **106 minors have Annual_Income recorded** (median ≈ R89,549) and **100 minors have a Credit_Score**.

## 3. Tenure inconsistencies
- **10 clients have `Customer_Tenure` greater than their `Age`** (i.e. they became customers before being born). These overlap with the negative-age rows plus other young clients.

## 4. Credit score out of range
- **1,084 credit scores fall outside the standard 300–850 range** (observed min = 26, max = 989).
- 369 credit scores are missing.

## 5. Product flag vs. balance contradictions
Per product, rows where the ownership flag disagrees with the balance column:

| Product | Flag = False but balance present | Flag = True but balance missing |
|---|---|---|
| Current_Account | 39 | 59 |
| Savings_Account | 46 | 53 |
| Credit_Card | 56 | 43 |
| Personal_Loan | 66 | 33 |
| Home_Loan | 76 | 22 |
| Vehicle_Loan | 71 | 28 |

## 6. Closed accounts with lingering balances
- All closed dates parse cleanly and fall in 2021–2024 (no future dates), and no account is simultaneously closed and flagged True.
- However, **15 closed accounts still carry a balance** (Current 3, Savings 5, Credit Card 2, Personal Loan 1, Home Loan 2, Vehicle Loan 2).

## 7. Suspicious balance caps / outliers
- **Home_Loan_Balance: 203 rows sit at exactly 2,800,000** (also the max) — looks like a hard cap or default fill value.
- **Vehicle_Loan_Balance: 175 rows at exactly 850,000** (the max) — same pattern.
- Credit_Card_Balance: 9 rows at exactly 220,000 (the max).
- Savings max of R1,651,710 belongs to a client with annual income of only R18,081 — income/balance mismatch.

## 8. Digital channel contradictions
- **Transactions recorded on channels the client is not registered for**: Online Banking 40 rows, Mobile App 31 rows, USSD 59 rows (txns > 0 while `_Registered` = False).

## 9. Income anomalies
- **2,310 of 2,346 'Unemployed' clients have positive Annual_Income** (median ≈ R21,273). Possibly grants, but inconsistent with the label.
- 163 incomes missing; none negative or zero.

## 11. Clients with no products at all
- **592 clients have every product flag set to False** — 'customers' holding nothing, which may be data loss or churned clients without closed dates.

## Notes on things that checked out fine
- No negative balances, tenures, transaction counts, or incomes.
- No unparseable or future closed dates.
- No leading/trailing whitespace or case inconsistencies in text columns; City↔Province pairings are geographically consistent.
