```python
import pandas as pd

# Load the file
df = pd.read_csv("SBSA DS Case Study 2027.csv")

# ---------------------------------------------------
# 1. Basic structure
# ---------------------------------------------------
print("Dataset shape:", df.shape)
print(f"Records: {df.shape[0]}")
print(f"Fields: {df.shape[1]}")

print("\nColumn names:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

# ---------------------------------------------------
# 2. Key numeric characteristics
# These are raw values for now; we will assess quality next.
# ---------------------------------------------------
key_numeric = [
    "Age",
    "Customer_Tenure",
    "Annual_Income",
    "Credit_Score"
]

print("\nKey numeric summary:")
print(
    df[key_numeric]
    .describe()
    .loc[["count", "mean", "50%", "min", "max"]]
    .T
    .round(2)
)

# ---------------------------------------------------
# 3. Product ownership
# ---------------------------------------------------
product_cols = [
    "Current_Account",
    "Savings_Account",
    "Credit_Card",
    "Personal_Loan",
    "Home_Loan",
    "Vehicle_Loan"
]

product_rates = (df[product_cols].mean() * 100).sort_values(ascending=False)

print("\nProduct holding rates (%):")
print(product_rates.round(1))

df["Number_of_Products"] = df[product_cols].sum(axis=1)

print("\nAverage products held per customer:")
print(round(df["Number_of_Products"].mean(), 2))

# ---------------------------------------------------
# 4. Digital-channel registration
# ---------------------------------------------------
digital_cols = [
    "Online_Banking_Registered",
    "Mobile_App_Registered",
    "USSD_Registered"
]

digital_rates = (df[digital_cols].mean() * 100).sort_values(ascending=False)

print("\nDigital registration rates (%):")
print(digital_rates.round(1))

df["Number_of_Digital_Channels"] = df[digital_cols].sum(axis=1)

print("\nAverage digital channels registered per customer:")
print(round(df["Number_of_Digital_Channels"].mean(), 2))

# ---------------------------------------------------
# 5. Geographic profile
# ---------------------------------------------------
print("\nProvince distribution (%):")
print((df["Province"].value_counts(normalize=True) * 100).round(1))

print("\nLocation type distribution (%):")
print((df["Location_Type"].value_counts(normalize=True) * 100).round(1))
```
